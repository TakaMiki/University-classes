```python
# 利用するライブラリをインポート
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# irisデータを pandas のデータフレームに読み込む (ヘッダがないので、下記のように指定)
iris_df = pd.read_csv('iris.csv')

print(iris_df.shape)  # irisデータのサイズを表示

print(iris_df.head())  # irisデータの先頭5行分を表示
```

```python
print('## iris data (all)')
print(iris_df.describe())  # データフレーム中のデータの要約統計量を取得
```

```python
# ペアプロットを表示 (hue の指定により species により色分け、diag_kind で対角線成分にヒストグラムを表示)
sns.pairplot(iris_df, hue='species', diag_kind='hist')
```

```python
print('## iris data (setosa)')
print(iris_df[iris_df['species']=='Iris-setosa'].describe())  # 品種 setosa のデータの要約統計量を取得
```

```python
# 品種 setosa を取り出し、ペアプロットを表示
sns.pairplot(iris_df[iris_df['species']=='Iris-setosa'])
```
