```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# 二項分布に従ったリストを作成 (p=0.5で10000回試行を1000回)
throw = 10000
p = 0.5
result_10000 = np.random.binomial(throw, p, size=1000)
# ヒストグラムを出力
plt.hist(result_10000, bins=10, ec='black')
plt.show()
# QQプロットを出力
stats.probplot(result_10000, dist="norm", plot=plt)
plt.show()
# Shapiro-Wilk 検定
print('throw=10000')
print(stats.shapiro(result_10000))
```

```python
# スライドで利用する図の作成
throw = 10
p = 0.5
result_10 = np.random.binomial(throw, p, size=1000)

throw = 100
p = 0.5
result_100 = np.random.binomial(throw, p, size=1000)

throw = 1000
p = 0.5
result_1000 = np.random.binomial(throw, p, size=1000)

throw = 10000
p = 0.5
result_10000 = np.random.binomial(throw, p, size=1000)

fig = plt.figure(figsize=(15, 7))

ax = fig.add_subplot(241)
ax.hist(result_10, bins=10, ec='black')
ax = fig.add_subplot(242)
ax.hist(result_100, bins=10, ec='black')
ax = fig.add_subplot(243)
ax.hist(result_1000, bins=10, ec='black')
ax = fig.add_subplot(244)
ax.hist(result_10000, bins=10, ec='black')

ax = fig.add_subplot(245)
stats.probplot(result_10, dist="norm", plot=plt)
ax = fig.add_subplot(246)
stats.probplot(result_100, dist="norm", plot=plt)
ax = fig.add_subplot(247)
stats.probplot(result_1000, dist="norm", plot=plt)
ax = fig.add_subplot(248)
stats.probplot(result_10000, dist="norm", plot=plt)

plt.show()

print('throw=10', stats.shapiro(result_10))
print('throw=100', stats.shapiro(result_100))
print('throw=1000', stats.shapiro(result_1000))
print('throw=10000', stats.shapiro(result_10000))
```
