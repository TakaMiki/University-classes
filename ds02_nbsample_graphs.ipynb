```python
import random
import statistics
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sklearn
from sklearn import datasets
import pandas as pd
import seaborn as sns

print('## ライブラリのバージョンを確認')
print('matplotlib:', matplotlib.__version__)

# 乱数発生用のシードを設定
random.seed(1)
```

```python
fig = plt.figure(figsize=(16, 6))

print('# 平均、中央値、最頻値の差の例 1')
x = []
for i in range(1000):
    x.append(int(random.normalvariate(500, 200)))
print('# 1')
print('mean', statistics.mean(x))
print('median', statistics.median(x))
print('mode', statistics.mode(x))
ax = fig.add_subplot(121)
ax.hist(x, bins=20, range=(0, 1000))

print('# 平均、中央値、最頻値の差の例 2')
x = []
for i in range(999):
    x.append(int(random.normalvariate(400, 100)))
x.append(100000)

print('# 2')
print('mean', statistics.mean(x))
print('median', statistics.median(x))
print('mode', statistics.mode(x))
ax = fig.add_subplot(122)
ax.hist(x, bins=20, range=(0, 1000))
plt.show()
```

```python
print('## 散布図 1')
x = []
y = []
for i in range(-100, 100):
    x.append(0.9 * i + random.uniform(-10, 10))
    y.append(0.9 * i + random.uniform(-10, 10))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.scatter(x=x,
           y=y,
           marker='o',
           c='blue')
plt.show()

print('r=', np.corrcoef(x, y))
```
