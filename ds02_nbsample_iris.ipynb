サンプルデータ iris を利用したデータ操作の基本の説明のための資料

```python
import random
import statistics
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sklearn
from sklearn import datasets
import pandas as pd
import seaborn as sns
import scipy

print('## ライブラリのバージョンを確認')
print('matplotlib:', matplotlib.__version__)
print('numpy:', np.__version__)
print('sklearn:', sklearn.__version__)
print('pandas:', pd.__version__)
print('seaborn:', sns.__version__)
print('scipy:', scipy.__version__)
```

```python
print('## サンプルデータ iris の確認')
## sklearn ライブラリに irisデータが収録されているのでそれを利用

iris = datasets.load_iris()  # サンプルデータ iris を変数 iris に代入

print('feature_names', iris.feature_names)  # サンプルデータの属性名の確認
print('data.shape', iris.data.shape)  # サンプルデータの属性値のサイズを確認
for i in range(3):
    print('    ', i, iris.data[i])  # サンプルデータを確認

print('target_names', iris.target_names)  # アヤメの品種名を確認
print('target.shape', iris.target.shape)  # アヤメの品種のサイズを確認
for i in range(3):
    print('    ', i, iris.target[i])  # サンプルデータを確認
```

```python
print('## サンプルデータ iris の中の品種 setosa のがく片データのグラフを作成')
iris_setosa = iris.data[iris.target==0]

print('# sepal length')
print('mean', statistics.mean(iris_setosa[:, 0]))
print('median', statistics.median(iris_setosa[:, 0]))
print('mode', scipy.stats.mode(iris_setosa[:, 0]))
print('variance', statistics.pvariance(iris_setosa[:, 0]))
print('stdev', statistics.pstdev(iris_setosa[:, 0]))
print('# sepal width')
print('mean', statistics.mean(iris_setosa[:, 1]))
print('median', statistics.median(iris_setosa[:, 1]))
print('mode', scipy.stats.mode(iris_setosa[:, 1]))
print('variance', statistics.pvariance(iris_setosa[:, 1]))
print('stdev', statistics.pstdev(iris_setosa[:, 1]))
print('# petal length')
print('mean', statistics.mean(iris_setosa[:, 2]))
print('median', statistics.median(iris_setosa[:, 2]))
print('mode', scipy.stats.mode(iris_setosa[:, 2]))
print('variance', statistics.pvariance(iris_setosa[:, 2]))
print('stdev', statistics.pstdev(iris_setosa[:, 2]))
print('# petal width')
print('mean', statistics.mean(iris_setosa[:, 3]))
print('median', statistics.median(iris_setosa[:, 3]))
print('mode', scipy.stats.mode(iris_setosa[:, 3]))
print('variance', statistics.pvariance(iris_setosa[:, 3]))
print('stdev', statistics.pstdev(iris_setosa[:, 3]))
```

```python
print('## サンプルデータ iris の中の品種 setosa のがく片データのグラフを作成')
iris_setosa = iris.data[iris.target==0]
iris_setosa_label = iris.target_names[0]

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

ax.scatter(x=iris_setosa[:, 0],
           y=iris_setosa[:, 1],
           label=iris_setosa_label,
           marker='o',
           c='blue')
ax.legend(loc='best', fontsize=14)

ax.set_title('iris (setosa) data (sepal length (cm) / sepal width (cm))', size=16)
ax.set_xlabel(iris.feature_names[0], size=14)
ax.set_ylabel(iris.feature_names[1], size=14)

plt.show()
```

```python
print('## サンプルデータ iris の setosa の各属性間の相関係数を計算')
iris_setosa = iris.data[iris.target==0]
correlation_matrix = np.corrcoef([iris_setosa[:, 0], iris_setosa[:, 1], iris_setosa[:, 2], iris_setosa[:, 3]])
print(correlation_matrix)

print('## サンプルデータ iris の setosa の各属性間の散布図行列を作成')
# skleanr の irisデータを pandas のデータフレームに変換し、seaborns の pairplot で散布図行列を作成
iris_setosa_df = pd.DataFrame(iris_setosa)
sns.pairplot(iris_setosa_df)
```
