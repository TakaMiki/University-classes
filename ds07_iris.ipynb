```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import openpyxl

# エクセルデータをデータフレームに読み込む
df = pd.read_excel('ds07_iris_setosa.xlsx')
print(df.head())
```

```python
# データフレームからX列とY列を取り出す
x = df[['X']]
y = df[['Y']]

# 単回帰分析を実施 (モデルを準備し、データをフィッテイング)
model_lr = LinearRegression()
model_lr.fit(x, y)

# 各数値を出力
print('回帰係数 b: %.3f' % model_lr.coef_)
print('切片    a: %.3f' % model_lr.intercept_)
print('y= %.3f + %.3fx' % (model_lr.intercept_, model_lr.coef_))
print('寄与率 R^2:', model_lr.score(x, y))
```

```python
import matplotlib.pyplot as plt

# 元データの散布図を作成し、求めた回帰直線を描画
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.plot(x, y, 'o')
ax.plot(x, model_lr.predict(x), linestyle="solid")
ax.set_title('iris (setosa) data', size=16)
ax.set_xlabel('sepal_length', size=14)
ax.set_ylabel('sepal_width', size=14)
plt.show()
```
