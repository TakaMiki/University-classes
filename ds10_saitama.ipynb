```python
# 重みなしグラフGから重み付きグラフG2を作成
G2 = nx.Graph()
g2_edges = []  # G2に登録するエッジのリストを準備
for edge in G.edges():
    # グラフGからエッジを取り出し、その頂点間の距離を計算
    dist = math.sqrt((pos[edge[0]][0]-pos[edge[1]][0])*(pos[edge[0]][0]-pos[edge[1]][0])
                    +(pos[edge[0]][1]-pos[edge[1]][1])*(pos[edge[0]][1]-pos[edge[1]][1]))
    print(edge, pos[edge[0]], pos[edge[1]], dist)
    g2_edges.append((edge[0], edge[1], dist))  # 2頂点とその間の距離をリストに登録
G2.add_weighted_edges_from(g2_edges)  # リストg2_edgesからグラフG2を作成
```

```python
print(G2)  # グラフの概要を表示
print(G2.size())  # グラフの大きさ(頂点数)を表示

plt.figure(figsize=(10, 6))  # グラフを表示
nx.draw_networkx(G2, pos, edge_color='darkblue', node_color='skyblue')
plt.show()
```

```python
MST = nx.minimum_spanning_tree(G2)  # グラフG2の最小全域木を作成
print(MST)
print(MST.size())

plt.figure(figsize=(10, 6))  # グラフG2をグレー、最小全域木を赤で表示
nx.draw_networkx(G2, pos, edge_color='lightgray', node_color='lightgray')
nx.draw_networkx(MST, pos, edge_color='red', node_color='salmon')
```

```python
# さいたま市(100)と本庄市(211)の間の最短経路を探索し、結果をリストで表示
SP_LIST = nx.shortest_path(G2, source=100, target=211, weight='weight')
print(SP_LIST)

# 最短経路上の市町村を登録したグラフSPを作成
SP = nx.Graph()
for i in range(len(SP_LIST)-1):
    nx.add_path(SP, [SP_LIST[i], SP_LIST[i+1]])

plt.figure(figsize=(10, 6))  # グラフG2をグレー、最短経路を赤で表示
nx.draw_networkx(G2, pos, edge_color='lightgray', node_color='lightgray')
nx.draw_networkx(SP, pos, edge_color='red', node_color='salmon')
```

```python
degree_centers = nx.degree_centrality(G2)  # 次数中心性を計算
print('degree centers') 
print(sorted(degree_centers.items(), key=lambda x: x[1], reverse=True)[:3])
close_centers = nx.closeness_centrality(G2)  # 近接中心性を計算
print('close centers')
print(sorted(close_centers.items(), key=lambda x: x[1], reverse=True)[:3])
between_centers = nx.betweenness_centrality(G2)  # 媒介中心性を計算
print('between centers')
print(sorted(between_centers.items(), key=lambda x: x[1], reverse=True)[:3])
pagerank = nx.pagerank(G2)  # ページランクを計算
print('page rank')
print(sorted(pagerank.items(), key=lambda x: x[1], reverse=True)[:3])
```
